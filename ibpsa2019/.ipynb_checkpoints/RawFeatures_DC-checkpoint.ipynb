```python
import pandas as pd
from datetime import datetime

df_dc = pd.read_csv('data/DGS_322_Buildings-15m-By_Building-DST-gap-filled-3-2-18-508pm.csv', 
                    parse_dates=[['Building ID', 'Unnamed: 1']], 
                    infer_datetime_format=True)

# get rid of temperature column
del df_dc['Unnamed: 2']

# update column names to match the row of building names
new_column_names = df_dc.iloc[0,:]
df_dc.columns = new_column_names

# get rid of rows with metadata and update index
df_dc = df_dc.drop([0,1,2], axis=0)
df_dc = df_dc.rename(columns = {'Building nan':'timestamp'})
df_dc.index = df_dc['timestamp'].astype('datetime64[ns]')
del df_dc['timestamp']

print(df_dc)
```

```python
# calculate amount of readings per day
import math
import datetime

totalDayReadings = pd.DataFrame() # initiliaze daaily reading summary dataset
firstColumn = True # aux boolean to insert timestamp as a column only once

for column in range(len(df_dc.columns)): # iterate through all buildings (column)
    currentBdg = pd.DataFrame(df_dc.iloc[:, column])

    # initialize the date with the very first timestamp of the dataset
    currentDate = datetime.datetime(currentBdg.index[0].year, currentBdg.index[0].month, currentBdg.index[0].day)

    buildingReadings = [] # list of cumulative readings in a day for all timestamps
    readingsDay = 0 # number of meter readings in a day
    dailyTimeStamp = [] # list of timestamps resample by day

    for index, row in currentBdg.iterrows(): # iterate through every instance for a giveng building
        meterValue = float(row[0])
        rowDate = datetime.datetime(index.year, index.month, index.day)
        # first verify we are still within the same day
        if rowDate == currentDate:
            if not math.isnan(meterValue):
                readingsDay += 1 # increase count if reading exists
            else:
                pass
        # if the days don't match, counts are done for that day
        else:
            if firstColumn: # only for the first column, keep track of timestamp
                dailyTimeStamp.append(currentDate) # append current date
            currentDate = rowDate # update curent timestamp
            buildingReadings.append(readingsDay) # save the cumulative count for previous day
            readingsDay = 0 # reset count of hourly readings

    if firstColumn: # only for the first column, insert the timestamp
        totalDayReadings['timestamp'] = dailyTimeStamp # insert timestamp as a table
        firstColumn = False # update aux boolen


    totalDayReadings[str(currentBdg.columns.values)] = buildingReadings # save readings as a column
    print("{} buildings so far".format(column + 1))
```

```python
df_dailyReadings= totalDayReadings.set_index('timestamp')
```

```python
import datetime
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)

data = [
    go.Heatmap(
        z=(df_dailyReadings.values).T,
        y=df_dailyReadings.columns.values,
        x=df_dailyReadings.index,
        colorscale='Viridis',
    )
]

axis_template = dict(
             showgrid = True, zeroline = False,
             linecolor = 'black', showticklabels = True,
             )

layout = go.Layout(
    title='Hourly Meter Data per day',
    autosize = True,
    xaxis = axis_template,
    yaxis = axis_template,
    )

fig = go.Figure(data=data, layout=layout)

plotly.offline.iplot(fig)

```

```python
from datetime import datetime

"""
Function that returns a random date, within an specified time period,
given a certain context
"""
def truncateDFbyContext(context, startDate, endDate, dataframe):
    dataframe = dataframe.fillna(0)
    # truncate time period
    dataframe_truncated = dataframe[(dataframe.index >= startDate) & (dataframe.index <= endDate)]

    # week day context
    if context == "week day":
        new_df = dataframe_truncated[(dataframe_truncated.index.weekday != 5) & 
                                     (dataframe_truncated.index.weekday != 6)]
    # weekend context
    elif context == "weekend":
        new_df = dataframe_truncated[(dataframe_truncated.index.weekday == 5) |
                                     (dataframe_truncated.index.weekday == 6)]

    return new_df

"""
Function to prepare the time series dataframe for proper analysis
1. Aggregates the hourley meter day per day
2. Delete days (rows) that were ignore based on context, thus have a value of 0
3. Filter out buildings without meter readings for the already selected time period
"""
def prepareDataframe(dataframe):
    # aggregate meter data by day
    dataframe = dataframe.resample('1D').sum()
    
    # delete the days with 0 values
    dataframe = dataframe[(dataframe.T != 0).any()]

    # replace 0.0 with NaN to drop columns with NaN
    dataframe = dataframe.replace(0.0, np.nan)
    dataframe = dataframe.dropna(axis=1)
    
    return dataframe.T
```

```python
import numpy as np

# Generate context matrices, dates chosen from the plot above, heatmap
startDate = df_dc.index[0]
endDate = df_dc.index[len(df_dc) - 1]

weekday_df = truncateDFbyContext("week day", startDate, endDate, df_dc)

# avoid having values as objects
weekday_df = weekday_df.astype(float)

# prepares the dataframe
weekday_df_prepared = prepareDataframe(weekday_df)

# test visualization
print(weekday_df_prepared)

# save the file as .csv
weekday_df_prepared.to_csv('data/weekdayContext_DC.csv')
```

```python
# subset of data 
weekend_df = truncateDFbyContext("weekend", startDate, endDate, df_dc)

# avoid having values as objects
weekend_df = weekend_df.astype(float)

# prepares the dataframe
weekend_df_prepared = prepareDataframe(weekend_df)

# test visualization
print(weekend_df_prepared)

# save the file as .csv
weekend_df_prepared.to_csv('data/weekendContext_DC.csv')
```
